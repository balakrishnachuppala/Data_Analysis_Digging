```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
df['FlightNumber']=np.where(df['FlightNumber'].isnull(),(df['FlightNumber'].shift()+10),df['FlightNumber'])
```

```python
df['FlightNumber']=df['FlightNumber'].astype(int)
```

```python
df
```

```python
df['From']=df['From_To'].apply(lambda x: (x.split('_')[0].capitalize()))
```

```python
df['To']=df['From_To'].apply(lambda x: (x.split('_')[1].capitalize()))
```

```python
del df['From_To']
```

```python
df
```

```python
df=df.join(pd.DataFrame(df['RecentDelays'].values.tolist()))
```

```python
df.rename(columns={0:'Delay_1',1:'Delay_2',2:'Delay_3'},inplace=True)
```

```python
del df['RecentDelays']
```
